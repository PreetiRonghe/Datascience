The main difference between `.loc` and `.iloc` in pandas is how they select data:

*   **`.loc` (label-based)**: This is primarily for selecting data by **label** (index names and column names). You can use a single label, a list of labels, or a slice of labels. When using slicing with `loc`, *both* the start and end labels are inclusive.

*   **`.iloc` (integer-location based)**: This is primarily for selecting data by **integer position** (from 0 to length-1). You can use a single integer, a list of integers, or a slice of integers. When using slicing with `iloc`, the start position is inclusive, and the end position is *exclusive* (just like standard Python slicing).

```python
# Example using iloc (integer-location based indexing)
print("df.iloc[0:3] (rows 0, 1, 2 - end is exclusive):")
display(df.iloc[0:3])

print("\ndf.iloc[:, 0:2] (all rows, columns 0, 1 - end is exclusive):")
display(df.iloc[:, 0:2])

print("\ndf.iloc[[0, 2], [1, 3]] (specific rows and columns by integer position):")
display(df.iloc[[0, 2], [1, 3]])
```

```python
# Example using loc (label-based indexing)
print("df.loc[0:2] (rows with index labels 0, 1, 2 - end is inclusive):")
display(df.loc[0:2])

print("\ndf.loc[:, 'company':'year'] (all rows, columns 'company' to 'year' - end is inclusive):")
display(df.loc[:, 'company':'year'])

print("\ndf.loc[[0, 2], ['model', 'price']] (specific rows and columns by labels):")
display(df.loc[[0, 2], ['model', 'price']])
```

```python
data={
    "month":["Jan","Feb","Mar","Apr"],
    "sales":[175000,125000,157000,132000]
}
df=pd.DataFrame(data)
print(df)
print("====================")
total_sales=df["sales"].sum()
print("Total Sales",total_sales)
```

```python
df['percentage']=(df['sales']/total_sales)*100
print(df)
#since we have created a df as percentage
```

```python
fig,ax1=plt.subplots()
ax1.bar(df['month'],df['sales'],color="blue")
ax1.set_xlabel("Months")
ax1.set_ylabel("Sales")
#ax1.set_title("Monthly Sales")

ax2=ax1.twinx()
ax2.plot(df['month'],df['percentage'],color="red",marker="o")
#ax2.set_ylabel("Percentage")

ax2=ax1.twinx()
ax2.plot(df['month'],df['percentage'],color="red",marker="o")
ax2.set_ylabel("Percentage of sales")

plt.title("Monthwise Sales")
plt.show()
```
