#String Datatype - Collection of characters or ste of character, that can be declared in Double Quotes or Single quotes

```python
#string
data2='Winter has started'
data1="Winter has started"
print(data1,data2)
print(type(data1),type(data2))
```

```python
data1="Winter has started "Everywhere" in India"
print(data1)
```

```python
data1='Winter has started "Everywhere" in India'
print(data1)
```

```python
data="winter's on of the best seasons we welcome it"
print(data)
```

```python
#if there is an apostrophe in a string use backslash with single quotes
data2='winter\'s on of the best seasons we welcome it'
print(data2)
#if there is an apostrophe in a string use backslash with double quotes
data3="winter\'s on of the best seasons we welcome it"
print(data3)
#if you want to pass multiple lines  use triple ''' or """"
data4="""winter's on of the best seasons we welcome it
it is my favourite season of all the seasons"""
print(data4)
#if you want to pass multiple lines  use triple ''' or """"
data5='''winter's on of the best seasons we welcome it
This is what we can do in Winter, Relax and Knit'''
print(data5)
data6="winter's on of the best seasons we welcome it and perfect for\"writers\""
print(data6)
```

```python
#independence day every alphabet will occupy one index and then space too followed by day
#the indexing starts from 0
#it fetches one character at a atime
data="independence"
print(data[0])
print(data[1])
#last character from string and second last character
print(data[-1])
print(data[-2])
```

```python
#Slicing fetches a series of character or subset of a string
data="Independence Day"
print(data[2:8])
print(data[::2])
print(data[::-1])
print(data[::])
print(data[2:8:2])
```

```python
#Uppercase lowerecase
data="Have a Good day!!"
print(data.upper())
print(data.lower())
print(data.title())
print(data.capitalize())
print(data.swapcase())
```
