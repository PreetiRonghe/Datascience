```python
import numpy as np
ar1=np.array([1,2,3,31,44,26])
print("The datatype of array",type(ar1))
print("==================")
print("The elements in array",ar1)
print("==================")
arr2=np.array([[1,2,3],[4,5,6]])
print("This is the actual array\n",arr2)
print("==================")
print("This will print the transpose of array\n",arr2.T)
```

```python
print("Shape",ar1.shape)
print("==================")
print("Number of elements",ar1.size)
print("==================")
print("Data Type",ar1.dtype)
print("==================")
print("size of each element in Byte",ar1.itemsize)
print("==================")
print("Dimension",ar1.ndim)
print("==================")
```

```python
print("==================")
l1=[12,11,1,5,67,84,34]
t1=(23,51,16,52)
s1={12,11,1,5,67,84,34}
print("==================")
ar1=np.asarray(l1)
print("==================")
print(ar1)
ar2=np.asarray(t1)
print("==================")
print(ar2)
#ar3=np.asarray(l1+t1)
ar3=np.asarray(l1+list(t1))
ar5=np.asarray(l1+list(s1))
print("==================")
print("Appending list and tuple",ar3)
print("==================")
ar4=np.asarray(t1,dtype=float)
print(ar4)
print("==================")
print("Appending list and Tuple",ar5)
```

```python
ar1=np.array([[1,3,6],[6,3,1]])
ar2=np.array([1,4,6])
print("Value of array 1\n",ar1)
print("Value of array 2\n",ar2)
print("==================")

print("Addition of aaray 1 and array 2")
print(np.add(ar1,ar2))
print("==================")

print("Subtraction")
print(np.subtract(ar1,ar2))
print("==================")

print("Multiplication")
print(np.multiply(ar1,ar2))
print("==================")

print("Division")
print(np.divide(ar1,ar2))
print("==================")

print("Log")
print(np.log(ar1))
print("==================")


print("Modulos / remainer")
print(np.mod(ar1,ar2))
print("==================")

print("EXPONENT / POWER")
print(np.power(ar1,ar2))
print("==================")
```
