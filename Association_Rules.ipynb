# Import libraries

```python
!pip install mlxtend
#moving pip install above.
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
titanic = pd.read_csv("Titanic.csv")
titanic.head()
```

# Pre-Processing
As the data is not in transaction formation
We are using transaction Encoder

```python
#We have to convert this data into Dummy variables
df=pd.get_dummies(titanic)
df.head()
df=pd.get_dummies(titanic,dtype=int)
df.head()
```

# Apriori Algorithm

```python
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
#try to execute this line to see the output frequent_itemsets = apriori(df, min_support=0.1, use_colnames=False)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
#min threshold change from 0.7 to 1
rules
rules.sort_values('lift',ascending = False).head(10)
#when lift is same check for confidence,Conviction shows that the consequent is dependent on anticedent
```

#####  An leverage value of 0 indicates independence. Range will be [-1 1]
A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.lift>1]
```
