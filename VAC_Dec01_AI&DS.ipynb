```python
#list is identified by []
#duplicate values are allowed
##heterogeneous datatype
data=[12,"Python",500,76.56, True]
data2=[12,12,121,121,12]
print(type(data))
print(type(data2))
print(data,data2)
#list is associated, starting from index 0
#indexing #Slicing #NEgative
```

```python
#ordered
data=["Apple","Cherry","Lime","Pineapple","Jackfruit","Kiwi","Dragon"]
print("First Element",data[0])
print("Last Element",data[-1])
print("2 to 6 element",data[1:6])
print("reversed",data[::-1])
```

```python
#append
data=["Apple","Cherry","Lime","Pineapple","Jackfruit","Kiwi","Dragon"]
print(data)
data.append("Mango")
print(data)
data.append("Orange")
print(data)
print("P--->",data[2])
print("D--->",data[5][0])
print(data)
```

```python
#insert , it inserts data based on the user defined index, this takes 2 values, the value that you want o insert and the position
data=["Apple","Cherry","Lime","Pineapple","Jackfruit","Kiwi","Dragon"]
print(data)
data.insert(1,"Mango")
print(data)
data.insert(3,"Orange")
print(data)
```

```python
#remove- elements can be deleted from random position. if the element is not present it will throw an erro. this method will delete the elemets by value
data=["Apple","Cherry","Lime","Pineapple","Jackfruit","Kiwi","Dragon"]
print(data)
data.remove("Apple")
print(data)
data.remove("Kiwi")
print(data)
data.remove("grapes")
```

```python
#pop will remove the last element by default or we can delete the element by the index value
data=["Apple","Cherry","Lime","Pineapple","Jackfruit","Kiwi","Dragon"]
print(data)
data.pop()
print(data)
data.pop(2)
print(data)
```

```python
#clear
# it will Remove all the elements from the list and the result will be an empty list
data=["Apple","Cherry","Lime","Pineapple","Jackfruit","Kiwi","Dragon"]
print(data)
data.clear()
print(data)
data.append("Grapes")
print(data)
```

```python
#CLEAR
data =[ 'html', 'css', 'javascipt', 'english', 'math', 'stats']
print(data)
data.clear()
print("EMPTY LIST",data)
data.append("Python")
print(data)
data.append("Java")
print(data)
```

```python
#del Keyword
data =[ 'html', 'css', 'javascipt', 'english', 'math', 'stats']
print(data)
print("DELETING THESE THREE ELEMENT --->",data[2:5])
print(data)
del data[2:5]
print(data)
```

```python
#if we delete the variable it means we are not able to call it again
print(data)
#del data delete a variable it self completely
print(data)
```
