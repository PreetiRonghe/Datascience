#special Characters


| Data Type    | Mutability | Description                                                               |
|--------------|------------|---------------------------------------------------------------------------|
| `string`     | Immutable  | Cannot be changed after creation. New string objects are created for modifications. |
| `list`       | Mutable    | Can be changed after creation (elements can be added, removed, or modified). |
| `tuple`      | Immutable  | Cannot be changed after creation. Elements are fixed.                     |
| `set`        | Mutable    | Can be changed after creation (elements can be added or removed).         |
| `dictionary` | Mutable    | Can be changed after creation (keys and values can be added, removed, or modified). |
