```python
my_dict = {'apple': 1, 'banana': 2}

# Case 1: Key exists
value1 = my_dict.setdefault('apple', 100)
print(f"After setdefault('apple', 100): my_dict = {my_dict}, returned value = {value1}")

# Case 2: Key does not exist
value2 = my_dict.setdefault('cherry', 3)
print(f"After setdefault('cherry', 3): my_dict = {my_dict}, returned value = {value2}")

# Case 3: Key does not exist, but no default value is provided (defaults to None)
value3 = my_dict.setdefault('grape')
print(f"After setdefault('grape'): my_dict = {my_dict}, returned value = {value3}")
```

```python
# Initialize a dictionary
graph = {}
# Add a key with a default empty list if it doesn't exist
graph.setdefault('A', []).append('B')
graph.setdefault('A', []).append('C')
graph.setdefault('B', []).append('D')
print(graph)
# Output: {'A': ['B', 'C'], 'B': ['D']}
```
