### Question 1: Basic Enumeration

Given a list of fruits: `fruits = ['apple', 'banana', 'cherry', 'date']`

Write a Python program using `enumerate` to print each fruit along with its index, in the format: `Index: X, Fruit: Y`

```python
fruit1 =enumerate(['apple', 'banana', 'cherry', 'date'])
#print(list(fruit1))
for index,fruit in enumerate(fruit1):
  print(f"Index:{index}, Fruits:{fruit}")
```

```python
fruits = ['apple','banana','cherry','date']
for index,fruit in enumerate(fruits):
  print(f"Index:{index}, Fruits:{fruit}")
```

### Question 2: Find Index of a Specific Element

Given a list of names: `names = ['Alice', 'Bob', 'Charlie', 'David', 'Eve']`

Write a Python program using `enumerate` to find and print the index of the name 'Charlie'. If 'Charlie' is not found, print a message indicating that.

```python
names = ['Alice', 'Bob', 'Charlie', 'David', 'Eve']
for index ,name in enumerate(names) :
  if(name == 'Charlie'):
    print(f'Charlie is at index {index}')
    break
else:
  print('Charlie not found')
```

### Question 3: Enumeration with a Starting Index

Given a list of tasks: `tasks = ['Review code', 'Write documentation', 'Deploy application']`

Write a Python program using `enumerate` to print each task with a starting index of 1 instead of 0, in the format: `Task #X: Y`

```python
tasks = ['Review code', 'Write documentation', 'Deploy application']
for index,task in enumerate(tasks,1):
  print(f"Task #{index}:{task}")
```

### Question 4: Filter Elements Based on Index

Given a list of numbers: `numbers = [10, 25, 30, 45, 50, 65, 70]`

Write a Python program using `enumerate` to create a new list containing only those numbers that are at an *even* index. Print the new list.

```python
numbers = [10, 25, 30, 45, 50, 65, 70]
even_numbers=[]
odd_numbers=[]
for index,number in enumerate(numbers):
  if(index%2==0):
    even_numbers.append(number)
  else:
    odd_numbers.append(number)
print(even_numbers)
print(odd_numbers)
```
