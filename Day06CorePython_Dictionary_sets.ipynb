```python
#There are multiple ways of adding elements to a new dictionary
#Way 1. Create a new Dictionary and then add the elements from it to old dictionary
add_capitals={"Japan":"Tokyo","Germany":"Berlin"}
#The Dictionary that you eant to update is written first followed by dictionary want to update from
capital_dict.update(add_capitals)
print(capital_dict)
```

```python
#Way 2- Use Update to add elements
capital_dict.update({"France":"Paris"})
print(capital_dict)
```

```python
#way 3- add multiple values
capital_dict.update({"USA":"Washington DC", "UK":"London"})
print(capital_dict)
```

```python
#pop() method removes the key value for which you have passed the key
capital_dict.pop("USA")
print(capital_dict)
```

```python
#in Dictionary you can only  change the values associated with the key and not the key
capital_dict["England"]="UK"
print(capital_dict)
```

```python
#loop Control statement
#Continue - when a particular continion is satisfied
#break - when we want to forcefully terminate the logical flow of statements, when the condition statement is satisfied the loop is broken and the execution stops
for i in range(1,11):
  print(i)
```

```python
#from ast import Continue
#The continue statement in Python is a control flow statement that is used inside loops (like for or while loops). When continue is encountered, it immediately stops the current iteration of the loop and moves on to the next iteration. Any code following the continue statement within that loop iteration will be skipped. It's useful when you want to skip certain parts of a loop's processing based on a condition, but still want the loop to proceed with subsequent iterations.
for i in range(1,11):
    if (i==5):
        continue
        #print("iteration has been skipped")
    print(i)
```

```python
for i in range(1,11):
    if (i==5):
        print("iteration has been skipped")
        continue
    print(i)
    if(i==7):
      break
```
