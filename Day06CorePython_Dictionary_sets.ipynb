```python
#There are multiple ways of adding elements to a new dictionary
#Way 1. Create a new Dictionary and then add the elements from it to old dictionary
add_capitals={"Japan":"Tokyo","Germany":"Berlin"}
#The Dictionary that you eant to update is written first followed by dictionary want to update from
capital_dict.update(add_capitals)
print(capital_dict)
```

```python
#Way 2- Use Update to add elements
capital_dict.update({"France":"Paris"})
print(capital_dict)
```

```python
#way 3- add multiple values
capital_dict.update({"USA":"Washington DC", "UK":"London"})
print(capital_dict)
```

```python
#pop() method removes the key value for which you have passed the key
capital_dict.pop("USA")
print(capital_dict)
```

```python
#in Dictionary you can only  change the values associated with the key and not the key
capital_dict["England"]="UK"
print(capital_dict)
```
