```python
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x*x, numbers))
print(squared_numbers)
```

```python
vowels=lambda n:True if n.lower() in'aeiou' else False
vowels('A')
```

```python
lstVowels=[]

for itr in "Thank you Universe":
  if vowels(itr)==True:
    lstVowels.append(itr)
print(lstVowels)
```

```python
#Filter--> takes two arguments
#Argument1--> A function
 #Argument2--> Iterable

lastVowels_filter = filter(vowels, "Thank you Universe")
print(list(lastVowels_filter))
```

```python
#recursive Function
#a function which calls itself
def factNum(n):
  if n==1:
    return n
  if n==0:
    return 0
  else:
    return n*factNum(n-1)

factNum(5)
```
