```python
vowels=lambda n:True if n.lower() in'aeiou' else False
vowels('A')
```

```python
lstVowels=[]

for itr in "Thank you Universe":
  if vowels(itr)==True:
    lstVowels.append(itr)
print(lstVowels)
```

```python
#Filter--> takes two arguments
#Argument1--> A function
 #Argument2--> Iterable

lastVowels_filter = filter(vowels, "Thank you Universe")
print(list(lastVowels_filter))
```

```python
#recursive Function
#a function which calls itself
def factNum(n):
  if n==1:
    return n
  if n==0:
    return 0
  else:
    return n*factNum(n-1)

factNum(5)
```

```python
#Logical Exception
#afunction to print out age
def age(age):
  try:
    if age < 0:
      raise Exception("Age can not be negative")
    return f'Age is {age}'
  except Exception as e:
    print(e)
    return -1
  else:
      print('No error at all')
  finally:
    print('This is always executed')

age(5)
```
