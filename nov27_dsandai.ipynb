```python
data = "Hello World!"
print(f"Original string: '{data}'")
print(f"ID of original string: {id(data)}")

# Using replace()
new_data_replace = data.replace("World", "Python")
print(f"String after replace(): '{new_data_replace}'")
print(f"ID of new string after replace(): {id(new_data_replace)}")
print(f"Original string after replace operation: '{data}' (still unchanged)")
print(f"ID of original string after replace operation: {id(data)}")

print("\n---")

# Using split()
list_data_split = data.split(" ")
print(f"List after split(): {list_data_split}")
print(f"Original string after split operation: '{data}' (still unchanged)")
print(f"ID of original string after split operation: {id(data)}")

# As you can see, the 'id' of the original string 'data' remains the same,
# while 'new_data_replace' is a completely new string object in memory.
# 'list_data_split' is a new list object, and the original string is not modified.
```

```python
# Using a list to store multiple values
values_of_a = []

# Assign 2 and add it to the list
a = 2
values_of_a.append(a)
print(f"Current value of a: {a}")
print(f"Values stored in list: {values_of_a}")

# Assign 3 and add it to the list
a = 3
values_of_a.append(a)
print(f"Current value of a: {a}")
print(f"Values stored in list: {values_of_a}")

# Assign 4 and add it to the list
a = 4
values_of_a.append(a)
print(f"Current value of a: {a}")
print(f"Values stored in list: {values_of_a}")

# Now you can access all the values from the list
print(f"The first value stored was: {values_of_a[0]}")
print(f"The final value stored in 'a' variable is: {a}")
```

This way, the variable `a` always holds its current, latest value, but you also have a record of its previous values stored in the `values_of_a` list.

## Understanding the `re` module (Regular Expressions)

The `re` module in Python allows you to work with regular expressions (often shortened to regex or regexes). Regular expressions are sequences of characters that define a search pattern. They are used for:

*   **Searching and finding patterns:** Locating specific text patterns within larger strings.
*   **Validating data:** Checking if a string conforms to a certain format (e.g., email addresses, phone numbers, passwords).
*   **Extracting information:** Pulling out specific pieces of data from unstructured text.
*   **Replacing text:** Substituting parts of a string that match a pattern with new text.
*   **Splitting strings:** Dividing a string into a list of substrings based on a pattern.

### Common `re` module functions:

*   `re.search(pattern, string)`: Scans through *string* looking for the first location where the *pattern* produces a match, and returns a match object, or `None` if no match is found.
*   `re.match(pattern, string)`: Similar to `search()`, but only matches at the *beginning* of the string.
*   `re.findall(pattern, string)`: Finds *all* non-overlapping matches of *pattern* in *string*, returning them as a list of strings.
*   `re.sub(pattern, repl, string)`: Replaces occurrences of *pattern* in *string* with *repl* (replacement string or function).
*   `re.split(pattern, string)`: Splits *string* by the occurrences of *pattern*.

### Basic Regular Expression Syntax Elements:

*   `.`: Matches any single character (except newline).
*   `*`: Matches 0 or more occurrences of the preceding character/group.
*   `+`: Matches 1 or more occurrences of the preceding character/group.
*   `?`: Matches 0 or 1 occurrence of the preceding character/group (makes it optional).
*   `[]`: Matches any one of the characters inside the brackets (e.g., `[aeiou]` for vowels).
*   `[^]`: Matches any character *not* inside the brackets (e.g., `[^0-9]` for non-digits).
*   `\d`: Matches any digit (0-9).
*   `\D`: Matches any non-digit character.
*   `\w`: Matches any alphanumeric character (letters, numbers, and underscore).
*   `\W`: Matches any non-alphanumeric character.
*   `\s`: Matches any whitespace character (space, tab, newline, etc.).
*   `\S`: Matches any non-whitespace character.
*   `^`: Matches the beginning of the string.
*   `$`: Matches the end of the string.
*   `|`: Acts as an OR operator (e.g., `cat|dog`).
*   `()`: Groups expressions and can be used for capturing substrings.

```python
import re

text = "The quick brown fox jumps over the lazy dog. My email is example@domain.com."

# --- Example 1: re.search() ---
# Find the first occurrence of 'fox'
match_fox = re.search(r"fox", text)
if match_fox:
    print(f"Found 'fox' starting at index: {match_fox.start()} and ending at index: {match_fox.end()}")
    print(f"Matched string: {match_fox.group()}")

print("\n---")

# Find a pattern: 'dog' followed by a period.
match_dog = re.search(r"dog\.", text)
if match_dog:
    print(f"Found 'dog.' at: {match_dog.group()}")

print("\n---")

# --- Example 2: re.findall() ---
# Find all words with 'o' or 'O'
all_o_words = re.findall(r"\b\w*o\w*\b", text, re.IGNORECASE)
print(f"Words containing 'o': {all_o_words}")

print("\n---")

# Find all numbers (if any)
numbers = re.findall(r"\d+", "There are 123 apples and 45 oranges.")
print(f"Numbers found: {numbers}")

print("\n---")

# --- Example 3: re.sub() ---
# Replace 'fox' with 'cat'
new_text_sub = re.sub(r"fox", "cat", text)
print(f"Text after replacement: {new_text_sub}")

print("\n---")

# Replace all whitespace with a single space
cleaned_text = re.sub(r"\s+", " ", "This   text   has   too   much   space.")
print(f"Cleaned text: '{cleaned_text}'")

print("\n---")

# --- Example 4: Extracting an email address ---
email_pattern = r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b"
email_match = re.search(email_pattern, text)
if email_match:
    print(f"Extracted email: {email_match.group()}")
```
