```python
#modify the function to return the total marks of each student

def student_marks(name,**marks):
  total = 0
  for sub,mark in marks.items():
    print(f'Name:{name}, Subject:{sub}, Marksscored:{mark}')
    total += mark
  return total

preeti_total = student_marks("Preeti",Maths=10,Science=20,Literature=30)
samru_total = student_marks("Samru",Maths=10,Science=20)

print(f' Preeti Total marks:{preeti_total}')
print(f' Samru Total marks:{samru_total}')
```

```python
def process_user_data(user_id, **user_details):
  print(f"Processing data for user ID: {user_id}")
  for key, value in user_details.items():
    print(f"  {key.replace('_', ' ').title()}: {value}")


print("Example 1:")
process_user_data(101, name="Alice", age=30, city="New York")

print("\nExample 2:")
process_user_data(102, name="Bob", occupation="Engineer", email="bob@example.com", status="active")
```

### Explanation of Keyword Variable Arguments (\*\*kwargs)

In Python, you can use `**kwargs` (which stands for keyword arguments) in a function definition to allow the function to accept an arbitrary number of keyword arguments. These arguments are collected into a dictionary within the function.

*   **`def process_user_data(user_id, **user_details):`**:
    *   `user_id` is a regular positional argument.
    *   `**user_details` captures all other keyword arguments passed to the function into a dictionary named `user_details`.

*   **Inside the function**:
    *   You can iterate through the `user_details` dictionary using `.items()` to access each keyword and its corresponding value.
    *   This allows for flexible function calls where you don't know in advance all the possible keyword arguments a user might provide.
