```python
#Special Characters
#\n
#\t
#\s
#gives next line
print('hello\nworld')
#gives a tab
print("hello\tworld")
#gives a space
print("hello\sworld")
#raw string
#if you use r it means dont consider any special characters that are mentioned in the string
print (r"hello\nworld")
#gives a double tab
print("hello\\world")
print("hello World's how are you")
# this will give an error print('hello World's how are you'')
print("hello World\'s how are you")
#print("hello\sworld")
time ="13:30"
num1=1
num2=2
print(f"num1 is {num1} and num2 is {num2}")
print(f"sum of {num1} and {num2} is {num1+num2}")
print(f'the time is {time}')


#\' - Single Quote
#\\ - Backslash
#\n - New Line
#\r - Carriage Return
#\t - Tab
#\b - Backspace
#\f - Form Feed
#\ooo - Octal value
#\xhh - Hex value
#\S-space
```

Input Function - used to get an input from the user. The value of the input always gets converted into a string. so the default datatype for input is string

```python
name = input("Enter your name")
print(name)
print(f'User name is {name}')
```

```python
age = input("Enter your age")
print(age)
print(f'User age is {age} and the data type is {type(age)}')
#As the inputs datatype is always string you will have to type cast this into an int first before you perform Typecasting, i.e convert this into int!
age_new = int( input("Enter your age"))
print(age_new)
print(f'User currrent age is {age} and your age next year is {age_new+1}')
```
