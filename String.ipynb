```python
#Find - Finds the first occurence of the string and it is case sensitive
str ="helloworld"
print(str.find('l'))
print(str.find('lo'))
print(str.find('la',0,5))
print(str.find('LD',0,5))
print(str.find('',2,4))
print(str.find('l',0,6))
```

```python
#Alternating Case: Without using built-in functions like replace() or join(), how would you change the case of every other character, starting with the first character in lowercase? For example, "helloworld" should become "hElLoWoRlD".


my_string ="helloworld"
result =''
for i in range(len(my_string)):
  if i%2==0:
    result+=my_string[i].lower()
  else:
    result+=my_string[i].upper()
print(result)
```

```python
#Consonants Only: How would you create a new string that contains only the consonants from "helloworld", maintaining their original order?

my_string ="helloworld"
result_C =''
result_v=''
for i in range(len(my_string)):
  if my_string[i] in ("a","e","i","o",'u'):
    result_C+=my_string[i].upper()
  else:
    result_v+=my_string[i].lower()
print(result_C)
print(result_v)
```
