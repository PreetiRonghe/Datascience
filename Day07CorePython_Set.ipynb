```python
#set Operation
#Union()
set1={"English","Hindi", "Marathi"}
set2={"Tamil","French","English"}
#set3=set1.union(set2)
set3=set2|set1
print("Output of Union:-->",set3)
#Intersection()
set1={"English","Hindi", "Marathi"}
set2={"Tamil","French","English","Hindi"}
set4=set1.intersection(set2)
print("Output of intersection:-->",set4)
#Difference()
set1={"English","Hindi", "Marathi"}
set2={"Tamil","French","English"}
set5=set1.difference(set2)
set6=set2.difference(set1)
print("Output of Set Difference:-->",set5)
print(set6)
#Symmetric_difference()
set1={"English","Hindi", "Marathi"}
set2={"Tamil","French","English"}
set7=set1.symmetric_difference(set2)
print("Output of Symmetric Difference:-->",set7)
#issubset()
set1={"English","Hindi", "Marathi"}
set2={"Tamil","French","English","Hindi"}
set8=set1.issubset(set2)
print("Output of is subset:-->",set8)
#intersection_update()
set1={"English","Hindi", "Marathi"}
set2={"Tamil","French","English","Hindi"}
set1.intersection_update(set2)
print("output of intersection_update:-->",set1)
```
