#sets
Unique Collection of elements

Think of them like a List of tuple
sets dont have duplicate values
Sets do not store any mutable items as members
sets are mutable


```python
#create a set
s1={1,2,3,4,5}
print(s1)
print(type(s1))
```

```python
#create a set
s2={1,2,2,3,4,5}
print(s2)
print(type(s2))
```

```python
#Sets are unordered
s3={5,4,3,2,1}
print(s3)
```

```python
s3.add(6)
print(s3)
```

```python
s3.pop()
print(s3)
```

```python
Set1 = {1,2,3,4,5}
Set2 = {4,5,6,7,8}
print(Set1.union(Set2))
```

```python
Set1 = {1,2,3,4,5}
Set2 = {4,5,6,7,8}
print(Set1.intersection(Set2))
print(Set2.intersection(Set1))
```

```python
Set1 = {1,2,3,4,5}
Set2 = {4,5,6,7,8}
print(Set1.difference(Set2))
print(Set2.difference(Set1))
```
