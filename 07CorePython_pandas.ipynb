## Understanding `.loc` and `.iloc` in Pandas

Pandas DataFrames provide two primary methods for indexing (selecting) data: `.loc` and `.iloc`.

### `.loc` (Label-based indexing)

`.loc` is primarily label-based. This means you pass the row and column labels (names) that you want to select.

*   **Row selection:** You can use a single label, a list of labels, or a slice of labels.
*   **Column selection:** You can use a single label, a list of labels, or a slice of labels.
*   It includes the end label when slicing.

```python
# Example using .loc on the 'udf' DataFrame
print("Selecting rows by label 'Harvard' and 'Yale' and columns 'SAT' and 'GradRate' using .loc:")
display(udf.loc[['Harvard', 'Yale'], ['SAT', 'GradRate']])

print("\nSelecting a slice of rows from 'Columbia' to 'Princeton' and all columns using .loc:")
display(udf.loc['Columbia':'Princeton', :])
```

### `.iloc` (Integer-location based indexing)

`.iloc` is primarily integer-location based. This means you pass integer positions (0-based index) for the rows and columns that you want to select.

*   **Row selection:** You can use a single integer, a list of integers, or a slice of integers.
*   **Column selection:** You can use a single integer, a list of integers, or a slice of integers.
*   It **excludes** the end integer when slicing (like standard Python slicing).

```python
# Example using .iloc on the 'udf' DataFrame
print("Selecting rows at integer positions 0 and 1, and columns at positions 0 and 2 using .iloc:")
display(udf.iloc[[0, 1], [0, 2]])

print("\nSelecting a slice of rows from index 2 to 5 (exclusive of 5) and all columns using .iloc:")
display(udf.iloc[2:5, :])
```

Handling Null Values

```python
import pandas as pd
import numpy as np
data=[
    [1,2,3,4,'M'],
    [np.nan,5,6,7,np.nan],
    [7,8,9,np.nan],
    [10,11,12,13,'M']
]
```

```python
df=pd.DataFrame(data,columns=['A','B','C','D','E'])
```

```python
df.fillna(0)
```

```python
df
```

```python
#Copy Datafram
df1=df.copy()
```

```python
#Replacing missing values with some values is called imputation
#Imputation-
#Numerical - mean or median
#categorical- mode
```

```python
avg_A=df['A'].mean()
avg_A
```

```python
avg_B=df['B'].mean()
#avg_B=round(avg_B,2)
avg_B
```

```python
avg_D =df['D'].mean()
avg_D
```

```python
df.shape
#returns rows and colums
```

```python
mode_E=df['E'].mode()
mode_E
```

```python
mode_E=df['E'].mode()[0]
mode_E
```

```python
df.fillna({'A':avg_A,'B':avg_B,'D':avg_D,'E':mode_E},inplace=True)
#We passed the fillvalue in the form of dictionary and inplace gives permamnet changes
```
