```python
from datetime import datetime
```

```python
data =datetime.now()
print(data)

# Example of using strftime to format the datetime object
formatted_data = data.strftime("%Y-%m-%d %H:%M:%S")
print(f"Formatted date: {formatted_data}")
```

```python
format_date=data.strftime("%d-%m-%Y -%H:%M:%S")
print(format_date)
```

```python
print("Current month and name", data.strftime("%m-%B"))
print(" week of year",data.strftime("%W"))
print("Current week number",data.strftime("%U"))
print("today,weekname",data.strftime("%A"))
print("Current Week Day No",data.strftime("%w"))
```

```python
data=datetime.now()
print(data)
print(data.year)
print(data.month)
print(data.day)
print(data.hour)
print(data.minute)
print(data.second)
```

```python
print("Current Date and time in DD-MM-YYYY - HH :MM :SS")
print(data.strftime("%d-%m-%Y - %H:%M:%S"))
print("Current Month NO and Month Name")
print(data.strftime("%m - %B"))
print("Current Week Number")
print(data.strftime("%U"))
print("Day Name")
print(data.strftime("%A"))
print("Current WEEK Day no")
print(data.strftime("%w"))
```

```python
#timedelta
from datetime import timedelta
#add 7 days to today
today=datetime.now()
future_date=today+timedelta(days=7)
print(future_date)
```

```python
#3 days before today
future_date=today-timedelta(days=3)
print(future_date)
```

```python
#difference between 2 dates
date1=datetime(2023,1,1)
date2=datetime.now()
diff=date2-date1
print(diff)
diff=date1-date2
print(diff)
```
