### Problem 1: Right-angled Triangle Pattern

**Goal**: Print a right-angled triangle pattern using asterisks, where each row has one more asterisk than the previous one.

**Example for `n = 5`:**
```
*
**
***
****
*****
```

````python
for i in range(1,6):
  print("*"*i)
````

````python
# Python code for Right-angled Triangle Pattern
def print_right_triangle(n):
    for i in range(1, n + 1):
        print("*" * i)

print_right_triangle(5)
````

### Problem 2: Inverted Right-angled Triangle Pattern

**Goal**: Print an inverted right-angled triangle pattern using asterisks, where each row has one less asterisk than the previous one.

**Example for `n = 5`:**
```
*****
****
***
**
*
```

````python
def print_inverted_triangle(n):
  for i in range(n,0,-1):
    print("*" * i)
print_inverted_triangle(5)
````

````python
# Python code for Inverted Right-angled Triangle Pattern
def print_inverted_right_triangle(n):
    for i in range(n, 0, -1):
        print("*" * i)

print_inverted_right_triangle(5)
````

### Problem 3: Pyramid Pattern

**Goal**: Print a pyramid (equilateral triangle) pattern using asterisks.

**Example for `n = 5`:**
```
    *
   ***
  *****
 *******
*********
```

````python
# Python code for Pyramid Pattern
def print_pyramid(n):
    for i in range(1, n + 1):
        # Print spaces
        print(" " * (n - i), end="")
        # Print asterisks
        print("*" * (2 * i - 1))

print_pyramid(5)
````

### Problem 4: Diamond Pattern

**Goal**: Print a diamond pattern using asterisks.

**Example for `n = 5` (number of rows for the top half):**
```
    *
   ***
  *****
 *******
*********
 *******
  *****
   ***
    *
```

````python
# Python code for Diamond Pattern
def print_diamond(n):
    # Upper half of the diamond
    for i in range(1, n + 1):
        print(" " * (n - i) + "*" * (2 * i - 1))

    # Lower half of the diamond
    for i in range(n - 1, 0, -1):
        print(" " * (n - i) + "*" * (2 * i - 1))

print_diamond(5)
````
