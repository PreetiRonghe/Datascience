```python
#Abstract Classes
#An abstract class in Python is a class that cannot be instantiated directly. It's meant to be inherited by other classes, and it can contain both concrete methods (methods with implementation) and abstract methods (methods declared without an implementation). Abstract methods act as a blueprint, forcing any concrete subclass to provide an implementation for them. This is often used to define a common interface for a group of related classes

from abc import ABC,abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def no_of_wheels(self):
        pass
    @abstractmethod
    def type_of_vehicle(self):
        pass
class lmvCar(Vehicle):
    def no_of_wheels(self):
        print("4")
    def type_of_vehicle(self):
        print("LMV")
class f1Car(Vehicle):
    def no_of_wheels(self):
        print("2")
    def type_of_vehicle(self):
        print("F1")
car_swifty=lmvCar()
car_swifty.no_of_wheels()
car_swifty.type_of_vehicle()
car_ferarri=f1Car()
car_ferarri.no_of_wheels()
car_ferarri.type_of_vehicle()
```

A class is like a blueprint for creating objects (a particular data structure), providing initial values for state (member variables or attributes), and implementations of behavior (member functions or methods).

```python
class Dog:
    # The __init__ method is a special method called a constructor.
    # It's automatically called when you create a new instance of the class.
    def __init__(self, name, breed):
        self.name = name  # Instance variable (attribute)
        self.breed = breed  # Instance variable (attribute)

    # A method is a function that belongs to an object.
    def bark(self):
        return f"{self.name} says Woof!"

    def get_info(self):
        return f"Name: {self.name}, Breed: {self.breed}"

# Instantiating the class (creating objects from the blueprint)
# We are creating two instances of the Dog class
dog1 = Dog("Buddy", "Golden Retriever")
dog2 = Dog("Lucy", "Beagle")

# Accessing attributes of the objects
print(f"Dog 1 Name: {dog1.name}")
print(f"Dog 2 Breed: {dog2.breed}")

# Calling methods on the objects
print(dog1.bark())
print(dog2.get_info())
```

```python
#multiple Inheritance with class fish
class fish:
    def __init__(self,first_name,last_name="fish",skeleton="bone",eyelids=False):
        self.first_name=first_name
        self.last_name=last_name
        self.skeleton=skeleton
        self.eyelids=eyelids
    def swim(self):
        print("The fish is swimming")
clown_fish=fish("clown")
print(clown_fish.first_name+" "+clown_fish.last_name)
print(clown_fish.skeleton)
print(clown_fish.eyelids)
clown_fish.swim()
```

```python
#Polymorphism - one attribute or method having different form
#+,*--> inbuilt method
#numbers--> they are addition, for strings they are concatenation

class Plus():
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def __add__(self,other):
        return self.a+other.a,self.b+other.b
new_plus=Plus(1,2)
new_plus.plus()
new_plus2=Plus(3,4)
new_plus2.plus()
print(new_plus+new_plus2)
```
