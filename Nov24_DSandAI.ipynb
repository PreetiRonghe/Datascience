```python
attempt = 0
password = "pree"
max_attempts = 3

while attempt < max_attempts:
    userpwd = input("Enter the password: ")
    if userpwd == password:
        print("Access successful!")
        break # Exit the loop on successful login
    else:
        print("Access denied!")
        attempt += 1 # Increment attempt only on failure
        if attempt < max_attempts:
            print(f"You have {max_attempts - attempt} attempts remaining.")
        else:
            print("You have reached the limit. Account locked.")

# This 'else' block for the while loop executes if the loop finishes without a 'break'
# meaning all attempts were exhausted.
else:
    if attempt == max_attempts:
        print("No attempts remaining. Please contact support.")
```
