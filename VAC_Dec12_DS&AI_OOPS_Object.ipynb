```python
class student_info():
  def __init__(self,name,age,location,creditamount,pin):
    self.name=name
    self.age=age
    self.location=location
    self.__creditamount=creditamount
    self.__pin=pin
    #2underscores will hide this variables
    def display(self):
      print("Student name is",self.name)
      print("Student age is",self.age)
      print("Student location is",self.location)
      print("Student credit amount is",self.__creditamount)
      print("Student pin is",self.__pin)
  def __del__(self):
      print("Object is destroyed")
```

```python
d1=student_info("Rahul",22,"Delhi",1000,1234)
print(d1.name)
print(d1.age)
print(d1.location)
#print(d1._student_info__creditamount)
#print(d1._student_info__pin)
#d1.display()
```

```python
#Name Mangling
#when you are forcefuly trying to access a hidden variable outdide of the class
#syntax objectname.1underscoreclassname2underscore_variablename
print(d1._student_info__creditamount)
print(d1._student_info__pin)
```

```python
#Child 1 Code2 inheritance
class iphone(Apple):
  def __init__(self,name,model,price,launch_year):
    self.name=name
    self.model=model
    self.price=price
    self.launch_year=launch_year
  def display_info(self):
    print(f"Product Name is {self.name}")
    print(f"Product model is {self.model}")
    print(f"Product price is {self.price}")
    print(f"Product year_launch is {self.launch_year}")
```

```python
#Child 1 Code3
p1= iphone("iPhone","17Pro",134900.00,2025)
p1.display_info()
p1.apple_info()
```

```python
#Child 2 Code2 inheritance with Products
class Products(Apple):
  def __init__(self,type,name,model,price,launch_year):
    self.type=type
    self.name=name
    self.model=model
    self.price=price
    self.launch_year=launch_year
  def display_info(self):
    print(f"Product Type is {self.type}")
    print(f"Product Name is {self.name}")
    print(f"Product model is {self.model}")
    print(f"Product price is {self.price}")
    print(f"Product year_launch is {self.launch_year}")
```

```python
#Child 2 Code 3
pr1=Products("Mobile","iPhone","17Pro",134900.00,2025)
pr1.display_info()
pr1.apple_info()
```

```python
class Person():
  def __init__(self,name,age,gender):
    self.name=name
    self.age=age
    self.gender=gender
  def person_info(self):
    print(f"Person name is {self.name}")
    print(f"Person age is {self.age}")
    print(f"Person gender is {self.gender}")
```

```python
class Company():
  def __init__(self,name,location,no_of_employees,turnover):
    self.name=name
    self.location=location
    self.no_of_employees=no_of_employees
    self.turnover=turnover
  def company_info(self):
    print(f"Company name is {self.name}")
    print(f"Company location is {self.location}")
    print(f"Company no_of_employees is {self.no_of_employees}")
    print(f"Company turnover is {self.turnover}")
```

```python
class Employee(Person,Company):
  def __init__(self, person_name, age, gender, company_name, location, no_of_employees, turnover, salary, designation):
    Person.__init__(self, person_name, age, gender)
    Company.__init__(self, company_name, location, no_of_employees, turnover)
    self.salary=salary
    self.designation=designation
  def employee_info(self):
    print(f"Employee salary is {self.salary}")
    print(f"Employee designation is {self.designation}")
```

```python
e1=Employee("Rohan",24,"MALE", "TATA","MUMBAI",5000,"43K",5000,"DATA SCIENTIST")
e1.person_info()
e1.company_info()
e1.employee_info()
```
