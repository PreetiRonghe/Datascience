```python
from math import exp
x=list(range(1,10))
y=[exp(i) for i in x]
#list comprehension
plt.plot(x,y)
plt.show()
```

```python
#plot multiple graph at a time: eg 12 graphs at a time-3*4 grid
from math import *
plt.figure(figsize=(10,10))
x=list(range(1,10))
for i in range(1,13):
    plt.subplot(3,4,i)
    y=[i*j for j in x]
    plt.plot(x,y)
plt.show()
```

```python
from math import *
plt.figure(figsize=(16,9))
x=list(range(-11,10))
y=[exp(i) for i in x]
plt.subplot(2,2,1)
plt.plot(x,y)
y=[i**2 for i in x]
plt.subplot(2,2,2)
plt.plot(x,y)
y=[i**3 for i in x]
plt.subplot(2,2,3)
plt.plot(x,y)
y=[sin(i)for i in x]
plt.subplot(2,2,4)
plt.plot(x,y)
plt.show()
```
