```python
from math import exp
x=list(range(1,10))
y=[exp(i) for i in x]
#list comprehension
plt.plot(x,y)
plt.show()
```

```python
#plot multiple graph at a time: eg 12 graphs at a time-3*4 grid
from math import *
plt.figure(figsize=(10,10))
x=list(range(1,10))
for i in range(1,13):
    plt.subplot(3,4,i)
    y=[i*j for j in x]
    plt.plot(x,y)
plt.show()
```

```python
from math import *
plt.figure(figsize=(16,9))
x=list(range(-11,10))
y=[exp(i) for i in x]
plt.subplot(2,2,1)
plt.plot(x,y)
y=[i**2 for i in x]
plt.subplot(2,2,2)
plt.plot(x,y)
y=[i**3 for i in x]
plt.subplot(2,2,3)
plt.plot(x,y)
y=[sin(i)for i in x]
plt.subplot(2,2,4)
plt.plot(x,y)
plt.show()
```

```python
import pandas as pd
df=pd.read_csv("mtcars.csv")
plt.scatter(df.mpg,df.hp)
plt.show()
```

```python
#histogram, Scatterplot and boxplot can be plotted on numerical columns
#Histogram single column is needed
#boxplot single attribute
#scatterplot 2 values are needed
#function used for histogram is plt.hist() all this is used from matplotlib
plt.hist(df.mpg)
plt.show()
plt.hist(df.mpg,edgecolor="red")
plt.show()
```

```python
#histogram
plt.hist(df.mpg,facecolor="yellow",edgecolor="green",bins=5)
plt.show()
```

```python
#inorder to find the extreme value in a graph we use boxplot
#it first plots the value in ascending order
#bargraph helps you to find utlier for categorical
#boxplot is used to plot outlier for numerical column
plt.boxplot(df.mpg)
plt.show()
```
