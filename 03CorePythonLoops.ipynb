````python
num=int(input("Enter the Number Greater than 500:"))
if(num>500):
  print("Number is Greater than 500")
  if(num==600):
    print("Number is 600")
  elif(num==650):
    print("Number is 650")
  elif(num==700):
    print("Number is 700")
  else:
    print("Number is not 600,650,700")
else:
  print("Number is not Greater than 500")
````

````python
print("Lets check your if concepts!")
num=int(input("Enter the number:"))
if num%3==0 and num%5==0:
  print("Number is Divisible by 3 and 5")
elif num%3==0:
  print("Number is Divisible by 3 only")
elif num%5==0:
  print("Number is Divisible by 5 only")
else:
  print("Number is not Divisible by 3 and 5")
print("Happy Coding!")
````
