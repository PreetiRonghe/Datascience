Two Important properties of OOP:
1. Inheritance
2. Polymorphism

under Mammals you can have a cats and dogs but they ahve their own peculiar  properties also



```python
#inheritance
class person:
  #define the constructor
  def __init__(self,name,age):
    #In constructor the first attribute is self create and initialize the object
    self.name=name
    self.age=age
    #define instance method
  def say_hello(self):
    print(f"Hello my name is {self.name}")
    #lets define another method
  def is_eligible(self):
    if self.age>=18:
      print("You are eligible to vote")
    else:
      print("You are not eligible to vote")
```

```python
#derived class
class astronaut(person):
  def __init__(self,name,age,mission):
    super().__init__(name,age)
    self.mission=mission
  def launch_satellite(self):
    print(f"{self.name} is launching a Mission{self.mission}")
```

```python
astronaut1=astronaut("John",25,"MissionMars")
```

```python
astronaut2=astronaut("tom",36,"Jupiter")
```

```python
astronaut1.launch_satellite()
```

```python
#create another derived class of person as dancer
class dancer(person):
  def __init__(self,name,age,style):
    super().__init__(name,age)# it will initialize name and age of the base class
    self.style=style
  def perform(self):
    print(f"{self.name} is performing {self.style} style")
#
```

```python
dancer1=dancer("Priya",25,"salsa")
```

```python
dancer1.perform()
```

```python
dancer2=dancer("Priya",25,"Bharatnatyam")
```

```python
dancer2.perform()
```

Private and public variable
private variable is defined using double underscore

```python
class person:
  def __init__(self,name,age):
    self.name=name #public attribute
    self.__age=age #private attribute
  def get_age(self):
    return self.__age
```

```python
class person:
  def __init__(self,name,age):
    self.name=name #public attribute
    self.__age=age #private attribute
  def get_age(self):
    return self.__age
  def set_age(self,age):
    #this function is defined to overwrite the private class
    self.__age=age
```

```python
person1=person("John",25)
```

Polymorphism - consider there are 2 classes and there are 2 seperate students with same name Amit in class.

Same name but they are different in 2 different classes

Sum is an example of Polymorphism

```python
class person:
  def __init__(self,name):
    self.name=name
  def say_hello(self):
    print(f"says Hello my name is {self.name} from parent class")
```

```python
class astronaut(person):
  def __init__(self,name,mission):
    super().__init__(name)
    self.mission=mission
  def say_hello(self):
    print(f"says Hello my name is {self.name} from child class")
```

```python
per=person('Tom')
```

```python
ast=astronaut('John','Mars')
```

```python
per.say_hello()
```

```python
ast.say_hello()
```
