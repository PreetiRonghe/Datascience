```python
#user defined classes
class Student():
  '''
  This is a draft version of the documentation. the document will be updated as and when the attributes or the definition of the class changes
  This is a student class, this is a blueprint or a template to create a student object. the student object will have some attributes like name, major, gpa, age, is_on_probation and we will keep adding new attributes to it.
  '''
  #Add attributes to the student
  def __init__(self,age, name, major, gpa, is_on_probation):
    self.name = name
    self.major = major
    self.gpa = gpa
    self.age = age
    self.is_on_probation = is_on_probation
  def student_says_hello(self):
    print("Hello")
  def print_student_details(self):
    print(f"Details of Students:")
    print(f"Name: {self.name}")
    print(f"Major: {self.major}")
    print(f"GPA: {self.gpa}")
    print(f"Age: {self.age}")
    print(f"Is on Probation: {self.is_on_probation}")
```

```python
# Instantiate the class
#Call the function
student1 = Student()
student1.student_says_hello()
```

```python
#Instantiate the class
student1= Student(25,"Pree","CompSci",7.8,"yes")
student1.print_student_details()
```

```python
#there are some inbuilt functions, which allows you to play around with the attributes of the class
#if you want to check , if a particular attribute ecists for an object or not
#hasattr()- returns true or false
#getattr() - returns the attribute value
#setattr()-changes the current value of an attribute with some other value
#delattr()- deletes the attribute of an object

print(hasattr(student1,"name"))
print(hasattr(student1,"age"))
print(getattr(student1,"name"))
print(getattr(student1,"age"))
setattr(student1,"age",26)
print(getattr(student1,"age"))
delattr(student1,"age")
print(hasattr(student1,"age"))
```

```python
#inheritance
class Vehicle():
    def __init__(self,make,model,year):
        self.make=make
        self.model=model
        self.year=year
    def print_vehicle_details(self):
        print(f"Vehicle Make: {self.make}")
        print(f"Vehicle Model: {self.model}")
        print(f"Vehicle Year: {self.year}")
```

```python
#Two more Vehicle class
#LMV ,4 , Wheels,800 CC
#CMV, 4,120CC
class LVM (Vehicle):
  pass
class CMV(Vehicle):
  pass
```

```python
#instantiate the LVM class
lvm=LVM('Toyota','Corolla',2020)
lvm.print_vehicle_details()
```

```python
#instantiate the CMV class
cmv=CMV('Toyota','Corolla',2020)
cmv.print_vehicle_details()
```

```python
class Locker():
  def __init__(self,locker_no,locker_type,locker_key_no):
    self.locker_no = locker_no
    self.locker_type = locker_type
    self.locker_key_no = locker_key_no
  def print_locker_details(self):
    print(f"Locker No: {self.locker_no}")
    print(f"Locker Type: {self.locker_type}")
    print(f"Locker Key No: {self.locker_key_no}")
  def is_eligible_for_a_locker(self,eligible_flag=False):
    print("Eligible for A Locker")
    eligible_flag = True
    return eligible_flag
```

```python
class personal_usage_locker(Locker):
  pass
```

```python
class Government_Locker(Locker):
  def __init__(self,binary_bits_no,secret_key,auth_person,ministry_name):
    self.ministry_name=ministry_name
    self.auth_person=auth_person
    self.secret_key=secret_key
    self.binary_bits_no=binary_bits_no
  def print_locker_details(self):
    print(f"Minstry Name: {self.ministry_name}")
    print(f"Auth Person: {self.auth_person}")
    print(f"Secret Key No: {self.secret_key}")
    print(f"Binary Bits No: {self.binary_bits_no}")
```

```python
gl= Government_Locker(2,'abc','Kree','Om')
gl.print_locker_details()
gl.is_eligible_for_a_locker(True)
```

```python
class High_Security_Locker(Locker):
  def __init__(self,locker_no,locker_type,locker_key_no,face_scan_id):
    self.face_scan_id=face_scan_id
    super().__init__(locker_no,locker_type,locker_key_no)
  def print_locker_details(self):
    print(f"Locker No: {self.locker_no}")
    print(f"Locker Type: {self.locker_type}")
    print(f"Locker Key No: {self.locker_key_no}")
    print(f"Face Scan ID: {self.face_scan_id}")
```

```python
hsl= High_Security_Locker(101,'hsl','BX087651','UID-456789')
hsl.print_locker_details()
hsl.is_eligible_for_a_locker(True)
```
