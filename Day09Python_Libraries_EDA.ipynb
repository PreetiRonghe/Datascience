Connecting Dataframes

```python
df1=pd.DataFrame([[1,2,3],[4,5,6]],columns=["A","B","C"])
df1
```

```python
df2=pd.DataFrame([[7,8,9],[10,11,12]],columns=["A","B","C"])
df2
```

```python
df3=pd.DataFrame([[1,2],[3,4]],columns=["C1","C2"])
df3
```

```python
#Concat for Records
#join for column
pd.concat([df1,df2],axis=0,ignore_index=True)
```

```python
df1.join(df3)
```

```python
#merging
d1=pd.DataFrame({"MovieID":[1,2,3,4,5],"Name":["Pathan","Jawan","Uri","Army","Don"]})
d1
```

```python
d2=pd.DataFrame({"MovieID":[1,3,4,5,6],"Rating":[3,4,5,3,2], "Member":['Amir','Salman','Ajay','Vicky','Ranbir']})
d2
```

```python
#merge
pd.merge(d1,d2,on="MovieID",how="left")
```

```python
pd.merge(d1,d2,on="MovieID",how="right")
```

```python
pd.merge(d1,d2,on="MovieID",how="inner")
```

```python
pd.merge(d1,d2,on="MovieID",how="outer")
```

Lambda Function : A Lambda Function is an anonymous function without nameand is used for short and simple operation


1.   Suppose we want to a function which will look at the years of experience
2. If it is more than 20 years we will say "very Experienced and if Less than 20 years, will say less experienced




```python
fn=lambda x:'Experience' if x>20 else 'Less Experienced'
```

```python
fn(10)
```

```python
sal['service'].apply (lambda x: 'More Experiences' if x>20 else 'Less experience')
```

```python
sal['status']=sal['service'].apply (lambda x: 'More Experiences' if x>20 else 'Less experience')
sal
```
