Using with Statement


*   Instead of Manually Opening and Closing the file, you can use the with statement, which will take care of automatically closing the file


*   This reduces the risk of the file Corruption and resource leakage




```python
with open("test.txt","r") as f:
  content=f.read()
  print(content)
```

```python
print("is the file closed?",f.closed)
```

Exception Handling
An exception should be handled and remaining part of code should be executed

```python
'''
basic syntax for exception handling is
except ExceptionType:
  try:
  except:
  finally:
'''
```

```python
n=10
try:
  res=n/0
except ZeroDivisionError:
  print("input a nonzero number for division")
```

```python
a=["food","Watermelon","Chickoo"]
print(a[10])
```

```python
#program executes successfully without error
try:
  a=["food","Watermelon","Chickoo"]
  print(a[10])
except:
  print("Invalid Index")
```

```python
#finally Block
'''
Exception Handling in Pythin
try:
  #code that
except:
except:
else:
finally:
  '''
```

```python
try:
  n=10
  res=100/n
except ZeroDivisionError:
  print("You cant divide by Zero!")
except ValueError:
  print("Enter a non zero Number!")
else:
  print("Result is",res)
finally:
  print("Executuion is Complete")
```

Raising Custom Exceptions
You can create your own exception handling mechanism by using raise


```python
def check_age(age):
  if age<18:
    raise ValueError("You must be 18 years or older")
  print("Access Granted")
try:
    check_age(15)
except ValueError as e:
    print (e)
```

```python
#let us handle above error using exceptional handling and display reciprocal
#program to print the reciprocal of even numbers
#i.e if user enters 6, display 1/6
try:
  num=int(input("Enter a number:"))
  assert num%2==0,"number is an even number"
  #as the above condition it true it moves ahead with the remaining code
except:
  print(num, "is not an Even Number!")
else:
  reciprocal=1/num
  print("The output of even number is",reciprocal)
```
